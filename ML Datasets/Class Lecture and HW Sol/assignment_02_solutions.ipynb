7. Use an iterative approach to find the best sun_pct threshold overall for the entire dataset.  Print out the best accuracy and the threshold that accuracy occurs at.

```python
def predict(row, thres):
    if row['sun_pct'] > thres:
        return 'foot'
    else:
        return 'car'
    
thres = 20
rate = .1

thres_l, acc_l = [], []

while thres <= 100:
    df['prediction'] = df.apply(lambda x: predict(x, thres), axis=1)
    df['correct_int'] = (df['prediction'] == df['commute_method']).astype(int)
    acc = df['correct_int'].mean()
    thres_l.append(thres)
    acc_l.append(acc)
    thres += rate

print(f'Threshold for best accuracy: {thres_l[np.argmax(acc_l)]}')
print(f'Best Accuracy: {max(acc_l)}')
```

8. Create a scatter plot of the accuracy per iteration of your loop. 

```python
plt.scatter(thres_l, acc_l)
plt.scatter(np.array(thres_l)[np.argwhere(acc_l==max(acc_l))], np.array(acc_l)[np.argwhere(acc_l==max(acc_l))], c='red')
plt.show()
```

10. Complete the same process as in question 7 above, but find the best sun_pct threshold *for each day of the week*.  Print out the thresholds for each day of the week and 

```python
def predict(row, thres):
    if row['sun_pct'] > thres:
        return 'foot'
    else:
        return 'car'

rate = .1

results_dict = {}

for dow in dows:
    print(dow)
    thres = 0
    thres_l, acc_l = [], []
    while thres <= 100:
        df_copy = df[df['day']==dow].copy()
        df_copy['prediction'] = df_copy.apply(lambda x: predict(x, thres), axis=1)
        df_copy['correct_int'] = (df_copy['prediction'] == df_copy['commute_method']).astype(int)
        acc = df_copy['correct_int'].mean()
        thres_l.append(thres)
        acc_l.append(acc)
        thres += rate
        results_dict[dow] = {'thres_l': thres_l, 'acc_l': acc_l}

thresh_dict = {k:v['thres_l'][np.argmax(v['acc_l'])] for k,v in results_dict.items()}
acc_dict = {k:v['acc_l'][np.argmax(v['acc_l'])] for k,v in results_dict.items()}
print(thresh_dict)
print(acc_dict)
```
