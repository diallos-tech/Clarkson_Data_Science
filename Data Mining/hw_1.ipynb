## i) Initial Datasets

```python
import numpy as np
import matplotlib.pyplot as plt

# a) Create a 95% confidence interval
dataset_i = np.array([43,63,53,50,58,44,53,53,52,14,50,43])

alpha = 0.05/2 
n_i = 12
x_bar_i = 50
std_i = 10
df_i = 11
sqt = np.sqrt(n_i)
print(f'a) The confidence interval for two-tail test is: ({x_bar_i} +- {(round(2.201*(std_i/sqt),2))})')

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = np.percentile(dataset_i, 25)
Q3 = np.percentile(dataset_i, 75)

# Calculate the IQR
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
filtered_data = dataset_i[(dataset_i >= lower_bound) & (dataset_i <= upper_bound)]

print('\nb) Boxplots')
# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot the boxplot with outliers
axes[0].boxplot(dataset_i, 0, 'gd')
axes[0].set_title('Boxplot with outliers in green')
axes[0].set_ylabel('Values')

# Plot the boxplot without outliers
axes[1].boxplot(filtered_data, 0)
axes[1].set_title('Boxplot without outliers')
axes[1].set_ylabel('Values')


print(f'c) The new confidence interval without the outlier is: ({round(filtered_data.mean(),2)} +- {round(2.228 * filtered_data.std()/np.sqrt(n_i - 1),2)})')
```
